```python
df['インデックス済']=data1
df['表示件数']=data2
```

```python
for i in range(len(df)):
    print(i)
    url = df.loc[i]["インデックス済"]
    print(url)
    html_text = requests.get(url).text
    soup = BeautifulSoup(html_text, 'html.parser')

    list_nav = soup.find(class_ = "list-page local-nav")
    lists = list_nav.find_all("li")

    # 口コミ件数
    review = lists[1]
    txt = review.text

    review_num = re.sub(r"\D", "", txt)

    # 商品一覧件数
    prod = lists[4]
    prod_txt = prod.text

    prod_num = re.sub(r"\D", "", prod_txt)

    #新商品

    try:

        non_txt = list_nav.find(class_="nothing").text

        #新商品
        if "新商品" in non_txt:
            new_prod = "×"
        else:
            new_prod = "●"

        #ランキング
        if "ランキング" in non_txt:
            ranking_prod = "×"
        else:
            ranking_prod = "●"
    except:
        pass

    # 他の口コミ
    other_review = soup.find(class_ = "list-type-5")
    other_review_num = len(other_review)

    # 口コミ数
    review_article = soup.find_all(class_ = "item-list-type-2")
    review_article = len(review_article)

    #x他の口コミの各件数下限
    other_review = soup.find(class_ = "list-type-5")
    num_list = []
    for lis in other_review.find_all("li"):
        txt = lis.text
        num = re.sub(r"\D", "", txt)
        try:
            num = int(num)
        except:
            pass
        num_list.append(num)
        
    min_review_num = min(num_list)

    # ページネーション

    pagination =soup.find(class_ = "pagination")
    pages = set(pagination.find_all("a"))
    next = set(pagination.find_all("a",class_="next"))

    pages = pages-next
    pages_num =len(pages)+1

    #データの追加
    print(url)

    df.loc[i]["口コミ件数"] =review_num
    df.loc[i]["新商品"] =new_prod
    df.loc[i]["ランキング"] =ranking_prod
    df.loc[i]["商品一覧件数"] =prod_num
    df.loc[i]["他の口コミ"] =other_review_num
    df.loc[i]["他の口コミの各件数下限"] =min_review_num
    df.loc[i]["口コミ数"] =review_article
    df.loc[i]["ページネーション"] =pages_num

    
```
